# 9.1

You observe a sequence of n normal random variables. The first is a standard normal: $X_1∼N(0,1)$. Then, the second random variable in the sequence has variance 1 and mean of the first random variable, so $X_2|X_1∼N(X_1,1)$. In general, $X_j|X_{j−1}∼N(X_{j−1},1)$.

a. Find $E(X_n)$ for n ≥ 2.

b. Find $Var(X_n)$.

### 9.1 Analytical Solution

(a)

A normal random variable can be redefined as: $N(u,z^2) = \mu + zN(0,1)$

## $E[X_2|X_1] ~ E[N(X_1,1)] = E[X_1 + N(0,1)] = E[X_1] + E[N(0,1)] = 0$  

Since we have $E[X_2|X_1]$, we can iteratively substitute these expectations all the way up to $X_n$. This will always yield zero.

(b)

Let's start with $X_2$:

## $Var(X_2) = E(Var(X_2 | X_1)) + Var(E(X_2 | X_1)) = 1 + Var(E[N(X_1,1)]) = 1 + 1 = 2$ 

The first part comes from the fact that the variance of each RV is always 1 by definition. The inner expectation of the second part evaluates to $X_1$, who's variance we know is 1. 

More generally we have:

## $Var(X_n) = E(Var(X_n | X_{n - 1})) + Var(E(X_n | X_{n - 1}))$

Let's think about $X_3$. Just as above, the first part will always be 1 by definition. For the right half, we have:

## $Var(E(X_3 | X_2)) = Var(E[N(X_2,1)]) = Var(X_2) = 2$

By iteratively substituting into the right hand side, the general solution looks like:

## $Var(X_n) = n$

```python
samps = 10000
n = 20

def sim_run():

    x = np.zeros(n)

    x[0] = np.random.normal(loc = 0, scale = 1, size = 1)

    for i in range(1,n):
        x[i] = np.random.normal(loc = x[i-1], scale = 1, size = 1)

    return x

sims = np.array([sim_run() for i in range(samps)])

sims = np.stack(sims, axis=0)

#

probs = pd.DataFrame({
    'n': np.arange(0,n)+1,
    'SampleVar': np.var(sims,axis=0),
    'AnaltVar': np.arange(0,n)+1
})


_ = plt.plot(probs['n'],
             probs['SampleVar'], 
             'ro')

_ = plt.plot(probs['n'],
            probs['AnaltVar'],
            'go--',
            linewidth = 2)

_ = plt.ylabel('Variance')
_ = plt.xlabel('X_n')

_ = plt.title('Analytical and Simulated Variance\nGreen: Analytical || Red: Estimated')
```

# 9.2

Let X∼N(0,1) and Y|X∼N(X,1). Find Cov(X,Y).

### 9.2 Analytical Solution 

## $Cov(X, Y) = E(XY) - E(X)E(Y) = E(XY) - 0E(Y) = E(XY)$ 

## $E(XY) = E[X * N(X,1)] = E[X * (X + N(0,1))] = E[X^2 + XN(0,1)] = E[X^2] = 1$ 

Since the Standard Normal is symmetrical around 0, scaling by another symmetrical Normal random variable will still peg the expectation to zero. 

```python
samps = 10000

x = np.random.normal(loc = 0, scale = 1, size = samps)
y = np.random.normal(loc = x, scale = 1, size = samps)

print('Estimated Cov(X,Y): %.3f' % np.cov(x,y)[0,1])

#

norm_prod = np.random.normal(loc = 0, scale = 1, size = samps)*np.random.normal(loc = 0, scale = 1, size = samps)

_= sns.distplot(norm_prod,
                hist = False, kde_kws={'clip': (-3.0, 3.0)})
_ = plt.title('Density of N(0,1) x N(0,1)\nSim Mean: %.2f' % np.mean(norm_prod))
```

# 9.3

You need to design a way to split \\$100 randomly among three people. ‘Random’ here means symmetrical: if Xi is the amount that the ith person receives, then X1,X2 and X3 must be i.i.d. Also, the support of X1 must be from 0 to 100; otherwise, we could simply assign each person a constant $33.33.

a. Consider the following scheme to randomly split the money: you draw a random value from 0 to 100 and give that amount to the first person, then you draw a random amount from 0 to the amount of money left ( \\$100 minus what you gave to the first person) and give that to the second person, etc. Show why this scheme violates the ‘symmetrical’ property that we want.

b. Consider the following scheme: generate three values, one for each person, from a Unif(0,1) r.v. Then, normalize the values (divide each value by the sum of the three values) and assign each person the corresponding proportional value out of \\$100 (i.e., if the first person has a normalized value of .4, give him \\$40). Show why this scheme results in the correct expectation for each person (i.e., the expectation satisfies the property of symmetry, and each person expects $33.33).

### 9.3 Analytical Solution

(a)  

First, let's rescale the total to 1 instead 100 so we can use the Standard Uniform. We already know that $E(X_1) = 0.5.$ Let's think about $E(X_2)$.

## $E(X_2) = E(E(X_2|X_1)) = E(\frac{1 - X_1}{2}) = E(\frac{1}{2} - \frac{X_1}{2}) = 0.25$  

We can already see this isn't symmetrical. For $X_3$, the expectation will depend on the sum of $X_1$ and $X_2$, since these will set the upper bound. This scheme actually doesn't even guarantee that the entire $100 will be allocated.

(b)  

Now that these distributions are completely independent we can simplify the math a bit:

## $100E(\frac{X_1}{X_1 + X_2 + X_3}) = E(\frac{0.5}{0.5 + 0.5 + 0.5}) = 33.33$

By symmetry, the expectations for $X_2$ and $X_3$ must also be 33.33.

```python
samps = 10000

# Part (a)

x_1 = np.random.uniform(low = 0, high = 1, size = samps)
x_2 = np.random.uniform(low = 0, high = 1-x_1, size = samps)
x_3 = np.random.uniform(low = 0, high = 1-x_1-x_2, size = samps)

print('Part (a)\nE[X1]: %.2f | E[X2]: %.2f | E[X3]: %.3f\n' % (np.mean(x_1),np.mean(x_2),np.mean(x_3)))

# Part (b)

x = np.random.uniform(low = 0, high = 1, size = (samps,3))

x_means = np.mean(x/np.sum(x, axis = 1)[:, np.newaxis], axis = 0)

print('Part (b)\nE[X1]: %.2f | E[X2]: %.2f | E[X3]: %.2f\n' % (x_means[0],x_means[1],x_means[1]))
```

# 9.4

Let X∼N(0,1) and Y=|X|. Find Corr(X,Y).

### 9.4 Analytical Solution

We can start off by looking at the covariance and then normalizing by the product of the standard deviations to arrive at the correlation.

## $Cov(X,|X|) = E[X|X|] - E[X]E[|X|]$

We know that the right hand side is zero since E[X] = 0. Let's think about the left hand side. X*|X| looks like $x^2$ above zero and $-x^2$ below zero. It is still symmetric around zero. That means that observing -4 and 4 is equally likely. Intuitively the expectation must also be zero since symmetry around zero is still preserved. Since covariance is zero, correlation must be zero as well. In other words, the absolute value operator erases any linear relationship, even though the variables are completely dependent.

```python
samps = 100000

x = np.random.normal(loc = 0, scale = 1, size = samps)
y = np.abs(x)

print('Est Corr(X,|X|) = %.2f' % np.corrcoef(x,y)[0,1])
```

# 9.5

CJ has X∼Pois(λ) chores to run, and will spend M∼Pois(λ) time at each chore. Time spent at each chore is independent of the number of chores and the time spent at other chores. Let Y be the total amount of time spent doing chores. Find E(Y).

## 9.5 Analytical Solution

The total time spent doing all chores is M x X. Since these two are independent, we can multiply their expectations:

## $E(Y) = E(E(Y|X,M)) = E(XM) = E(X)E(M) = \lambda^2$

```python
samps = 10000
lamb = 3

x = np.random.poisson(lam = lamb, size = samps)

# Each draw is a vector of times to complete the i chores simulated by x
y = [np.sum(np.random.poisson(lam = lamb, size = x[i])) for i in range(samps)]

print('Sim E[Y]: %.2f' % np.mean(y))
```

# 9.6 + 9.7

Brandon is a cell. During his life cycle, he has Pois(λ) offspring before he dies.

a. Each of his descendants have i.i.d. Pois(λ) offspring distributions (i.e., like Brandon they independently have Pois(λ) children before they die). Let Xn be the size of the nth generation, and let Brandon be the 0th generation (so X0=1, since Brandon is just 1 cell, and X1 is the number of children that Brandon has, or the first generation). Find E(Xn).

b. Discuss for what values of λ the generation mean goes to infinity as n grows, and for what values the generation mean goes to 0

c. Find $Var(X_n)$

### 9.6 Analytical Solution

(a)

Let's traverse the first two generations:  

## $E(X_1|X_0) = E(Pois(\lambda)) = \lambda$

## $E(X_2|X_1) = E(\sum_{n=1}^{X_1} Pois(\lambda))$  

We know that on average there will be $\lambda$ offspring in the first generation and that each of these will on average have their own $\lambda$ offspring. Therefore, we expect $\lambda^2$ offspring. This pattern continues generally and we get that $E(X_n) = \lambda^n$.

(b)

Since we've established that $E(X_n) = \lambda^n$, it is clear that $\lambda < 1$ will cause this expectation to collapse to zero as n grows. $\lambda > 1$ will cause population growth without an upper bound. Keep in mind that this is true on average and in the limit as n grows.

(c) [Problem 9.7]

Let's try to work through a few examples manually. It is useful to remember that for a Poisson RV, Mean = Variance = $\lambda$.

## $Var(X_1) = E(Var(X_1|X_0)) + Var(E(X_1|X_0)) = E(\lambda X_0) + Var(\lambda X_0)$

Since there is no variability around $X_0$ (it is always 1), we're only left with the left hand side, which simplifies to $\lambda$.

### $Var(X_2) = E(Var(X_2|X_1)) + Var(E(X_2|X_1)) = E(\lambda X_1) + Var(\lambda X_1) = \lambda^2 + \lambda^2*\lambda = \lambda^2 + \lambda^3$

### $Var(X_3) = E(Var(X_3|X_2)) + Var(E(X_3|X_2)) = E(\lambda X_2) + Var(\lambda X_2) = \lambda^3 + \lambda^2*(\lambda^2 + \lambda^3) = \lambda^3 + \lambda^4 + \lambda^5$  

A pattern is emerging, the solution is just the sum of (n-1) successive powers starting at $\lambda^n$. The exception to this rule is $X_0$, where the variance is zero.

## $Var(X_n) = \sum_{i=n}^{2n-1} \lambda^i$

```python
samps = 10000

lamb = 2
n = 6

def sim_cycle(lamb,n):

    x = np.zeros(n+1,dtype=int)
    x[0] = 1

    for i in range(1,n+1):

        if x[i-1] == 0:
            break

        x[i] = np.sum(np.random.poisson(lam = lamb, size = x[i-1]))
        
    return x
    
sims = np.array([sim_cycle(lamb = lamb, n = n) for i in range(samps)])

sims = np.stack(sims, axis=0)

#

sim_means = np.mean(sims, axis=0)

sim_var = np.var(sims, axis=0)

#

def analt_var(lamb,n):
    return np.sum([lamb**i for i in range(n,2*n)])

#

probs = pd.DataFrame({
    'n': np.arange(1,n+1),
    'SampleVar': sim_var[1:len(sim_var)],
    'AnaltVar': [analt_var(lamb = lamb, n = i) for i in np.arange(1,n+1)]
})


_ = plt.plot(probs['n'],
             probs['SampleVar'], 
             'ro')

_ = plt.plot(probs['n'],
            probs['AnaltVar'],
            'go--',
            linewidth = 2)

_ = plt.ylabel('Variance')
_ = plt.xlabel('X_n')

_ = plt.title('Analytical and Simulated Variance\nGreen: Analytical || Red: Estimated')
```

# 9.8

Each year, the top (Men and Women) collegiate basketball teams in the NCAA square off in a massive, single elimination tournament. The tournament is colloquially known as “March Madness”, and, despite recent expansion to include ‘play-in’ games, it can be thought of as a 64-team tournament. Teams are ‘seeded’ (i.e., assigned a seed, 1 to 16) based on their performance during the season. Lower seed values are better (i.e., 1 is the best, 16 is the worst) and the teams are paired in the first round based on seeds (i.e., they are paired such that the seeds total to 17: each 1 seed plays a 16 seed, each 10 seed plays a 7 seed, etc.).

Of late, the UConn Huskies have had highly successful tournaments. The Men’s Team have won championships in 2011 and 2014 (as well as previously in 1999 and 2004) and the Women’s Team, considered by many the greatest collegiate team in the nation (across all sports), won 4 straight championships from 2013 to 2016, as well as 7 championships from 1995 to 2010.

Of course, in this tournament, it does not make sense to assume that teams are equal; in fact, they are seeded based on their ability. Consider this common model to assess win probability of for a random match up. Let *a* be the seed of the first team, and *b* be the seed of the second team, and let *p* be the probability that the first team wins. We can model  p  with a Beta prior such that  p∼Beta(b,a) . Based on this prior, find the probability that the first team wins, and, based on this probability, explain why this is a reasonable choice for a prior (i.e., consider how the probability changes as  a  changes relative to  b ).

### 9.8 Analytical Solution

The setup essentially just creates a Beta distribution with a and b swapped. This means the expected probability team *a* wins is just the modified expectation: 

 ## $E(\text{a win}) = \frac{b}{a+b}$

As b increases relative to a, the expected probability of *a* winning increases. The estimate is always bounded between 0 and 1. This is a reasonable prior choice.

# 9.9

Let X and Y be i.i.d. N(0,1) and Z be a random variable such that it takes on the value X (the value that X crystallizes to) or the value Y (the value that Y crystallizes to) with equal probabilities (recall we saw a similar structure in Chapter @ref(covariance-and-correlation), where we showed that the vector (X,Y,Z) is not Multivariate Normal). Find Cov(X,Z).

### 9.9 Analytical Solution

Let's first substitute into the standard covariance definition.

## $Cov(X,Z) = E[XZ] - E[X]E[Z] = E[XZ]$

We know the right hand side is zero since $E[X] = 0$. We can expand the $E[XZ]$ using the Law of Total Expectation.

### $E[XZ] = E[XZ|Z=X]P(Z=X) + E[XZ|Z=Y]P(Z=Y) = 0.5E[X^2] + 0.5E[X]E[Y] = 0.5$

```python
samps = 10000

x = np.random.normal(loc = 0, scale = 1, size = samps)
y = np.random.normal(loc = 0, scale = 1, size = samps)

z = [(x[i],y[i])[int(np.random.binomial(n = 1, p = 0.5, size = 1))] for i in range(samps)]

print('Est Cov(X,Z): %.2f' % np.cov(x,z)[0,1])
```

# 9.10

A stoplight in town toggles from red to green (no yellow). The times for the ‘toggles’ (switching from the current color to the other color) are distributed according to a Poisson process with rate parameter λ. If you drive through the stoplight at a random time during the day, what is your expected wait time at the light?

### 9.10 Analytical Solution

If we arrive at the light when its green, there is not wait time necessary. If we arrive when the light is red, we need to wait for it to turn green. We can use the Law of Total Expectation here.

### $E[wait] = E[wait | light=green]P(light=green) + E[wait | light=red]P(light=red) = 0.5E[Exp(1/\lambda)] = \frac{1}{2\lambda}$

```python
samps = 10000
lamb = 3

state = [1 if i % 2 == 0 else 0 for i in range(samps)]
time = np.random.exponential(scale = 1/lamb, size = samps)

print('E[wait time]: %.2f' % np.mean(time * state))
```

# 9.11

Dollar bills are the base currency in the United States. Bills are used widely in 6 denominations: \\$1, \\$5, \\$10, \\$20, \\$50, \\$100 (the \\$2 still exists, but is not widely used). Imagine that you randomly select one of these denominational values from \\$5 to \\$100 (i.e., \\$10) and withdraw it from your bank. On average, how many withdrawals must you make to withdraw at least \\$15?

### 9.11 Analytical Solution

Let's explicitly count the number of ways we can get a certain number of draws:

3 draws only works when we have 2 consecutive 5s and anything else  

2 draws works in two cases  
* 5 and anything except 5  
* 10 and anything else  
    
1 draw works if we get {20,50,100}  

Weighing each scenario by the probability gives us:  

## $\frac{1}{5}^2 * 3 + (\frac{1}{5}\frac{4}{5} + \frac{1}{5}) * 2 + \frac{3}{5} = \frac{36}{25} = 1.44$

```python
samps = 10000

bills = np.array([5,10,20,50,100],dtype=int)

sim_mean = np.mean([np.argmax(np.cumsum(np.random.choice(bills, replace = True, size = len(bills))) >= 15) + 1 for i in range(samps)])

print('Siml Mean: %.3f' % sim_mean)
print('Real Mean: %.3f' % ((1/5)**2 * 3 + (1/5 * 4/5 + 1/5) * 2 + 3/5 * 1))
```
