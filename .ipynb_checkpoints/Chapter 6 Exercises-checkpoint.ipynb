## $f(x,y) = \int_{0}^{1} \int_{0}^{1} 1 \;dx \;dy = 1$  

b. Marginalizing out Y is pretty simple since the bounds don't depend on either variable:  

## $f(x) = \int_{0}^{1} f(x,y) \;dy = \int_{0}^{1} 1 \;dy = 1$

c + d. It's easy to see that observing any y-value has no effect on the potential x-values we can see. If we see first that y = 0.6, this gives us no information on what x may be. More formally if $f(x) = f(x|y)$ then we have independence.  

## $f(x|y) = \frac{f(x,y)}{f(y)} = 1/1 = f(x)$

# 6.2

Let X∼N(0,1) and Y|X∼N(X,1). Find the joint PDF of X and Y.

### 6.2 Analytical Solution

We know that:  

## $f(y|x) = \frac{f(x,y)}{f(x)},\; f(x,y) = f(x)f(y|x)$  

Which gives us:  

## $f(x,y) = \frac{1}{\sqrt{2 \pi}} e^{-\frac{x^2}{2}} \frac{1}{\sqrt{2 \pi}} e^{-\frac{(y - x)^2}{2}} = \frac{e^{\frac{-2x^2+2xy-y^2}{2}}}{2\pi}$  

We can confirm this integrates to 1 using Wolfram Alpha:

http://www.wolframalpha.com/input/?i=integration+%5B%2F%2Fmath:e%5E((-2x%5E2%2B2xy-y%5E2)%2F2)%2F(2*pi)%2F%2F%5D+%5B%2F%2Fmath:dy+dx%2F%2F%5D+for+x+from+%5B%2F%2Fmath:-inf%2F%2F%5D+to+%5B%2F%2Fmath:inf%2F%2F%5D+for+y+from+%5B%2F%2Fmath:-inf%2F%2F%5D+to+%5B%2F%2Fmath:inf%2F%2F%5D

```python
# Analytical PDF
def pdf_62(x,y):
    return np.exp((-2*x**2 + 2*x*y - y**2)/2)/(2*np.pi)

# Generating PDF values for points in (-4,4)
samps = 100000

points = np.random.rand(samps,2)*8 - 4

dens = [pdf_62(points[i,0],points[i,1]) for i in range(samps)]

#

_ = plt.ylabel('Y-value')
_ = plt.xlabel('X-value')
_ = plt.title('Joint PDF when X∼N(0,1) and Y|X∼N(X,1)')
_ = plt.scatter(points[:,0], points[:,1], c = dens, edgecolors='none')
```

# 6.3

You roll a fair, six-sided die X times, where X∼Pois(λ). Let Y be the number of 6’s that you roll. Find P(Y>0); you may leave your answer in terms of λ.

### 6.3 Analytical Solution

This is very similar to the Chicken & Egg problem, since X is Poisson and Y is Binomial. The distribution of the successes is:  

### $Success \sim Pois(\lambda p)$

Which in this case translates to:  

### $Num(6) \sim Pois(\lambda/6) = \frac{\lambda^ye^{-\lambda}}{y!}$  

We only care about excluding the part of the PMF where y=0. This gives us:  

## $P(Y > 0) = 1 - e^{-\lambda/6}$

```python
# Simulates probability that sample greater than zero
def greater_zero(lamb, samps = 10000):

    X = np.random.poisson(lam = lamb, size = samps)
    Y = np.random.binomial(n = X, p = 1/6, size = samps)

    return np.mean(Y > 0)

#

comp = pd.DataFrame({
    'Lambda': np.arange(2,22),
    'Prob_Sim': [greater_zero(lamb = l) for l in range(2,22)],
    'Prob_Analt': [1 - np.exp(-l/6) for l in range(2,22)]
})

#

_ = plt.plot(comp['Lambda'],
             comp['Prob_Analt'])

_ = plt.plot(comp['Lambda'],
             comp['Prob_Sim'], 
             'ro')

#

_ = plt.title('Prob Y > 0\nBlue = Actual || Red = Simulated')

_ = plt.ylabel('Prob Y>0')
_ = plt.xlabel('Lambda')
```

# 6.4

There are n quarters. You flip them all (they are fair) and place them on a table. You then select m≤n quarters at random. Let X be the number of quarters that show heads on the table, and Y be the number of quarters that show heads that you select. Find the Joint PMF of X and Y.

### 6.4 Analytical Solution

There are three levels of uncertainty coming from this scenario: the number of heads from the initial flips (X), the number of coins we choose (m), and how many of them are heads (Y).  

X is easy to think about, it matches the standard description of the Binomial distribution, Bin(n,p).  

The number of coins we choose, m, is bounded above by n and below by zero. Since we're completely picking at random, this boils down to a discrete uniform probability for each choice in [0,n]. Its important to note that there are n+1 options here since we are including zero.  

When we arrive at the sampling step for Y we have: m total objects, x desirable objects, and n-x undesirable objects. We want to find the probability of a certain number of successful draws. This matches the description of the Hypergeometric distribution, HGeom(x,n-x,m). This is dependent on the previous two random variables, X and m.

In this specific case, since X and m are independent we can apply the following logic:  

## $f(x,m,y) = f(x,m)f(y|x,m) = f(x)f(m)f(y|x,m)$  

## $f(x,m,y) = \binom{n}{x}p^x(1-p)^{(n-x)} \; * \; \frac{1}{n+1}  \; * \; \frac{\binom{x}{y}\binom{n-x}{m-y}}{\binom{n}{m}}$

```python
# Simulating joint probability numbers
samps = 100000
n = 4
p = 1/2

# Simulates a single draw for Y
def sim_Y(n, x, m):
    
    coins = np.concatenate((np.ones(x),np.zeros(n - x)))
    
    samp = np.random.choice(coins, size = m, replace = False)

    return np.sum(samp)
    
#

X = np.random.binomial(n = n, p = p, size = samps)

m = np.random.randint(low = 0, high = n+1, size = samps)

Y = [sim_Y(n = n, x = X[i], m = m[i]) for i in range(samps)]
 
#

samp = pd.DataFrame({
    'm': m,
    'X': X,
    'Y': Y
})

# Aggregate sample counts
res = samp.groupby(['X','m','Y']).size()/samps * 100

# Compute analytical number for a single configuration
def head_prob_analt(n,p,x,m,y):
    
    bin_part = binom(n,x) * p**x * (1-p)**(n-x)
    
    hgeom_part = binom(x,y) * binom(n-x,m-y) / binom(n,m)
    
    m_prob = 1/(n+1)
    
    return (bin_part*hgeom_part*m_prob)

# Compare analytical and simulated numbers
analt_res = [100*round(head_prob_analt(n = 4, 
                           p = 1/2, 
                           x = res.index[i][0], 
                           m = res.index[i][1], 
                           y = res.index[i][2]),4) for i in range(res.size)]

final_res = pd.DataFrame({
    'Sim_Prob': res,
    'Analt_Prob': analt_res
})

final_res
```

# 6.5

Imagine a standard clock with the hours 1 to 12. The small hand starts at a random hour and moves counterclockwise (i.e., from 4 to 3) or clockwise (i.e., from 10 to 11) with equal probabilities. Let $X_t$ be the hour that the small hand is at in round t (i.e., after the hand has moved t times). Find the joint PMF of $X_t$ and $X_{t+1}$ for t≥1.

### 6.5 Analytical Solution

There are two flavors of this problem. The simpler one is where the small hand starts at a truly random hour, we do not condition on a specific starting point. This gives us a uniform 1/12 probability of being at any hour, $X_t$. Since only one time period has passed between $X_t$ and $X_{t+1}$, the small hand can either be one position to the left or to the right of its place at $X_t$. Here $p = 1/2$ since the hand moves in either direction with equal probability. This leaves us with:  

## $P(X_t = x_t , X_{t + 1} = x_{t + 1}) = P(X_t = x_t) P(X_{t + 1} = x_{t + 1} | X_t = x_t) = 1/12 * 1/2$  

A more complicated version of this problem is where: the probability of moving to the right is $p$ and we are given a specific starting position $s$. The $P(X_{t + 1} = x_{t + 1} | X_t = x_t)$ part from above would be simple in this scenario. It would be $p$ probability of landing to the right of $X_t$ and $(1-p)$ probability of landing to the left. The tricky part here is $P(X_t = x_t)$.  

The number of $right$ moves follows a Binomial distribution parameterized by: $t$ - the number of time periods we are sampling and $p$ - the probability of a success. Let's say that $t = 2$ and $p = 1/2$. We are interested in finding the $net$ (right - left) movement of all of the configurations:  

$\begin{array}{rr} \hline
\text{Probability} &0.25 &0.5 &0.25 \\ \hline
\text{Right Moves} &0 &1 &2 \\ \hline
\text{Left Moves}  &2 &1 &0 \\ \hline
\text{Net Moves} &-2 &0 &2 \\ \hline
\end{array}$  

The general pattern is that $\text{Net Moves} = (2r - t)$ where $r$ is the number of right moves coming from $R \sim Bin(t,p)$. To get the resulting position on the clock, we can add the starting position to the net moves, $Position = s + 2r - t$. So if we started at 6pm, observed 2 time steps, and none of them were to the right - our ending position would be 4pm. 

There is one small issue with this logic - what happens if this equation results in something that falls outside of [1..12]? Since the support for this distribution must remain on the clock hours, we need to adjust for this. The system of equations below adjusts the support so that it always maps back to the numbers on the clock. It basically "wraps" $(s + 2r - t)$ around the clock.  

$\begin{array}{rr} \hline
If \;(s + 2r - t) \in 12n &Position = 12 \\ \hline
If \;(s + 2r - t) \in 12n+1 &Position = 1 \\ \hline
If \;(s + 2r - t)\in 12n+2 &Position = 2 \\ \hline
If \;(s + 2r - t) \in 12n+3 &Position = 3 \\ \hline
If \;(s + 2r - t) \in 12n+4 &Position = 4 \\ \hline
If \;(s + 2r - t) \in 12n+5 &Position = 5 \\ \hline
If \;(s + 2r - t) \in 12n+6 &Position = 6 \\ \hline
If \;(s + 2r - t) \in 12n+7 &Position = 7 \\ \hline
If \;(s + 2r - t) \in 12n+8 &Position = 8 \\ \hline
If \;(s + 2r - t) \in 12n+9 &Position = 9 \\ \hline
If \;(s + 2r - t) \in 12n+10 &Position = 10 \\ \hline
If \;(s + 2r - t) \in 12n+11 &Position = 11 \\ \hline
\end{array}$  

Where:  
$n \in \mathbb{Z}$  
$0 \leq r \leq t$  
$R ~ \sim Bin(t,p)$  

Iterating over $r$, weighing by the probabilities of $R$, and applying the equations above will result in the PMF of $X_t$.

```python
# Simulation based solution
samps = 10000
s = 6
t = 31
p = 6/7

pos_sims_raw = np.array([np.sum(np.random.choice([-1,1], size = t, p = [1-p,p])) for i in range(samps)])

pos_sims = np.mod(s + pos_sims_raw,12)

pos_sims[pos_sims == 0] = 12

#

tmp = pd.DataFrame({'pos': pos_sims}).groupby('pos').size()

est_prob = pd.DataFrame({
    'pos': list(tmp.index),
    'est prob': tmp.values/samps
})

est_prob['est prob'] = np.round(100*est_prob['est prob'],2)
```

```python
# Analytical solution
steps = np.arange(t+1)

pos = np.mod(s + 2*steps - t,12)

pos[pos == 0] = 12

probs = sps.binom.pmf(k = steps, n = t, p = p)

real_prob = pd.DataFrame({
    'pos': pos,
    'real prob': probs
}).groupby(['pos']).sum().reset_index()

real_prob['real prob'] = np.round(100*real_prob['real prob'],2)

real_prob.merge(est_prob, on = 'pos', how = 'left')
```

# 6.6

Imagine a standard Rubix cube: 6 faces, each with a separate color, each face colored with 9 stickers (for a total of 54 stickers on the cube). You peel off n≤54 random stickers. Let R be the number of red stickers that you get, and Y be the number of yellow stickers. Find the Joint PMF of R and Y.

### 6.6 Analytical Solution 

It is important to note here that R and Y *alone* follow a Hypergeometric distribution. We have $w = 9$ desired objects, $b = 45$ undesired object, $n$ samples, and some specific number of successes $x$ we're looking to see in that sample. However, together they cannot follow this logic:

## $P(R = r, Y = y) = P(R = r|Y = y)P(Y = y)$

### $P(R = r, Y = y) = PHGeom(w = 9, b = 45, n, x = r) * PHGeom(w = 9, b = 45 - r, n = n-r, x = y)$

The reason is that when picking Y, we did not condition that the failures are not R. The Hypergeometric logic above just gives the probability the failures are not Y. It is actually just easier to start from scratch using basic combinatorics.

There are (54 choose n) ways to construct groups of $n$ stickers. Of the 9 possible R we need only r. Of the 9 possible Y we need only y. The rest of the sample must necessarily be neither R nor Y, there are 36 other options left. Order doesn't matter here so we are free to use the binomial coefficient everywhere:  

## $P(R = r, Y = y)  = \frac{\binom{9}{r}\binom{9}{y}\binom{36}{n-r-y}}{\binom{54}{n}}$

```python
# Simulated solution
n = 4
nsamps = 10000

#

stickers = np.concatenate((
    np.ones(9)*1,
    np.ones(9)*2,
    np.ones(9)*3,
    np.ones(9)*4,
    np.ones(9)*5,
    np.ones(9)*6
))

def sticker_sample(n):

    # sample = np.random.choice(stickers, size = n, replace = False)

    np.random.shuffle(stickers)
    
    sample = stickers[0:n]
    
    return np.array((np.sum(sample == 1),np.sum(sample == 2)))
    
#

samples = np.array([sticker_sample(n = n) for i in range(nsamps)])

est_prob = pd.DataFrame({
    'R': samples[:,0],
    'Y': samples[:,1]
}).groupby(['R','Y']).size().to_frame('Est P(R,Y)')/nsamps * 100

est_prob['Est P(R,Y)'] = np.round(est_prob['Est P(R,Y)'],2)
```

```python
# Analytical solution
def sticker_prob(n, R, Y):
    
    count_R = binom(9,R)
    count_Y = binom(9,Y)
    count_other = binom(36,n-R-Y)
    
    count_total = binom(54,n)
    
    return count_R*count_Y*count_other/count_total
    
# All possible Red and Yellow sticker combinations
nums = np.arange(n+1)

combs = np.array([np.array((x,y)) for x in nums for y in nums if x+y<=n])

# Getting probs over support
probs = np.array([sticker_prob(n = n, R = combs[i,0], Y = combs[i,1]) for i in range(combs.shape[0])])

# Formatting
real_prob = pd.DataFrame({
    'R': combs[:,0],
    'Y': combs[:,1],
    'Real P(R,Y)': np.round(probs*100,2)
},
columns = ['R','Y','Real P(R,Y)']).groupby(['R','Y']).mean()

#

pd.concat([real_prob,est_prob], keys = ['R','Y'], axis = 1)
```

# 6.7

Let X, Y and Z be continuous random variables with the joint PDF given by $f(x,y,z)=x^2+y^2+z^2$, where the support of each random variable is the interval from 0 to 1. Find f(x), f(y) and f(z), and check that they are all valid PDFs.

### 6.7 Analytical Solution

To isolate f(x) we need to marginalize out the other variables, y and z. This practically translates into integrating them out one at a time:  

## $f(x) = \int_{0}^1 \int_{0}^1 f(x, y, z) dz dy = \int_{0}^1 \int_{0}^1 x^2 + y^2 + z^2 dy dz$

Let's start off with the inner integral with respect to y:  

## $\int_{0}^1 x^2 + y^2 + z^2 dy = yx^2 + \frac{y^3}{3} + yz^2 \Big|_0^1 = x^2 + \frac{1}{3} + z^2$  

Finally, we can move on to the outer integral:  

## $\int_{0}^1 x^2 + \frac{1}{3} + z^2 dz = zx^2 + \frac{z}{3} + \frac{z^3}{3} \Big|_0^1 = x^2 + \frac{2}{3} = f(x)$  

By symmetry, the other variables must follow the same logic. This leaves us with:  

## $f(x) = x^2 + \frac{2}{3}$  
## $f(y) = y^2 + \frac{2}{3}$  
## $f(z) = z^2 + \frac{2}{3}$  

If we integrate any function we get a total area of 1, which makes it a valid PDF:  

## $\int_{0}^1 x^2 + \frac{2}{3} = \frac{x^3}{3} + \frac{2x}{3} \Big|_0^1 = 1$

# 6.8

Let X,Y be i.i.d. N(0,1). Find $E[(X+Y)^2]$ using 2-D LoTUS. You can leave your answer as an unsimplified integral.

### 6.8 Analytical Solution

First, we can expand out the inside of the expectation and then run it through the result. We know that for a standard normal variable $E[X] = 0$ and $E[X^2] = 1$.

## $E[(X+Y)^2] = E[X^2 + 2XY + Y^2] = E[X^2 + Y^2]$  

Next we can apply the 2-D LoTUS:  

## $E[X^2 + Y^2] = \frac{1}{2\pi} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (x^2 + y^2) e^{-x^2/2}e^{-y^2/2}dx dy = 2$  

Without computing the integral, we can notice that $E[X^2 + Y^2] = E[X^2] + E[Y^2] = 2$. This works since X and Y are independent.

```python
print('Simulated estimate for E[(X+Y)^2] = %s' % (
    np.round(np.mean(np.random.normal(size = 10000)**2 + np.random.normal(size = 10000)**2),2)
))
```

a. In the case where $λ_1 = λ_2$, what is the probability that the bus arrives first?

b. In the case where $λ_1$ and $λ_2$ are not necessarily equal, what is the probability that the bus arrives first? Provide some intuition (i.e., discuss how this answer changes with $λ_1$ and $λ_2$ and say why this is the case).

c. Now imagine that CJ is also waiting for a plane that has wait time $Z \sim Pois(λ_3)$, independently of X and Y. What is the probability that this plane arrives first (i.e., before the bus and car), in the general case where the parameters of X, Y and Z are not necessarily equal?

```python
# Parts (a) and (b)
nsamps = 100000
lamb1 = 1 # X
lamb2 = 2 # Y

X = np.random.poisson(lam = lamb1, size = nsamps)
Y = np.random.poisson(lam = lamb2, size = nsamps)

print('Simulated prob from %s sims: %s' % (nsamps,np.round(np.mean(X < Y),6)))

# Approximate solution
def inner_sum(lamb1,y):
    return (np.exp(-lamb1)*np.sum([(lamb1**x)/factorial(x) for x in range(y)]))

def full_sum_approx(lamb1,lamb2,limit):
    return (np.sum([np.exp(-lamb2)*lamb2**y/factorial(y) * inner_sum(lamb1,y) for y in range(1,limit)]))
    
print('Approximate analytical prob: %s' % np.round(full_sum_approx(lamb1,lamb2,100),6))
```

```python
# Part (c)
nsamps = 100000
lamb1 = 3 # X
lamb2 = 4 # Y
lamb3 = 1 # Z

X = np.random.poisson(lam = lamb1, size = nsamps)
Y = np.random.poisson(lam = lamb2, size = nsamps)
Z = np.random.poisson(lam = lamb3, size = nsamps)

print('Simulated prob from %s sims: %s' % (nsamps,np.round(np.mean((Z < X) * (Z < Y)),6)))

def inner_sum2(lamb,z):
    return (1 - np.exp(-lamb)*np.sum([lamb**x/factorial(x) for x in range(z+1)]))

def full_sum_approx2(lamb1,lamb2,lamb3,limit):
    return (np.exp(-lamb3) * np.sum([lamb3**z/factorial(z)*inner_sum2(lamb1,z)*inner_sum2(lamb2,z) for z in range(100)]))

print('Approximate analytical prob: %s' % np.round(full_sum_approx2(lamb1,lamb2,lamb3,100),6))
```

### 6.9 Analytical Solution

a + b) First, since a Poisson random variable is discrete, P(X < Y) won't be 1/2 even if both variables follow the same distribution. This is shown in the solution to problem 4.2.

Let's imagine that the car arrives in 4 hours, Y = 4. If the bus (X) arrives first, it has to arrive within [0,1,2,3] hours. If the car arrives in 10 hours then the bus has to arrive in [0..9] hours to be first. This pattern continues across the entire support of Y just without the 0, $[1,\infty)$. We are excluding zero since if the car arrives immediately, there is no way for the bus to arrive first. The general solution looks like: (Probability Y takes on some value) x (Probability that X's value is below that)  

## $\sum_{y=1}^{\infty} P(Y = y)*P(X < y)$

## $P(X < Y) = \sum_{y=1}^{\infty} [\frac{e^{-\lambda_2} \lambda_2^y}{y!} \sum_{x=0}^{y-1} \frac{e^{-\lambda_1} \lambda_1^x}{x!}]$  

c) Now we're looking for P(Z < X) & P(Z < Y). We can break down the problem like this:  

(Probability Z takes on some value) x (Probability X's value is greater than Z) x (Probability Y's value is greater than Z)  

We can multiply the probabilities here since they are independent. Knowing that the bus has arrived gives you no information about the car or plane.  

## $\sum_{z=0}^{\infty} \frac{e^{-\lambda_3} \lambda_3^z}{z!} * (1 - \sum_{x=0}^{z} \frac{e^{-\lambda_1} \lambda_1^x}{x!}) * (1 - \sum_{y=0}^{z} \frac{e^{-\lambda_2} \lambda_2^y}{y!})$
